Using Neural Network

```python
x = torch.tensor(5.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

Forward Propagation


```python
#forward prop
z = w * x + b
z
```

```python
#sigmoid act func
y_pred = torch.sigmoid(z)
y_pred
```

```python
#loss
def binary_cross_entropy_loss(prediction, target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction, epsilon, 1-epsilon)  #eitber 0 or 1
  return -(target *  torch.log(prediction) + (1-target) * torch.log(1-prediction))
```

```python
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

Backward Propagation

```python
loss.backward()
```

```python
x.grad
x
```

```python
b.grad
b
```

PyTorch Autograd is PyTorch's automatic differentiation engine. It enables the computation of gradients for all operations in a computation graph, which is crucial for training neural networks through backpropagation. It helps by automatically calculating derivatives, freeing developers from manual gradient computation and allowing for more complex model architectures and dynamic computation graphs.
