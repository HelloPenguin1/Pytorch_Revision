## Start of NN Module Intro

### Testing NN Module on a single layer

```python
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__() #invoking the nn constructor
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    #z = wx + b
    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

```python
# create dataset
features = torch.rand(10,5)

model = Model(features.shape[1])

model(features) #automatically forward function runs
```

### Testing NN module on multiple hidden layers

```python
class Model2(nn.Module):
  def __init__(self, num_features):
    super().__init__() #invoking the nn constructor
    self.linear1 = nn.Linear(num_features, 3)  # n features to 3 neurons thru linear layer
    self.relu = nn.ReLU()  #activation function
    self.linear2 = nn.Linear(3, 1)  # 3 neurons to one thru second linear layer
    self.sigmoid = nn.Sigmoid() # final activation for prediction


  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
```

### Simplifying forward operation with Sequential container

```python
class Model2(nn.Module):
  def __init__(self, num_features):

    super().__init__() #invoking the nn constructor

    self.network = nn.Sequntial(
    nn.Linear(num_features, 3),  # n features to 3 neurons thru linear layer
    nn.ReLU(),  #activation function
    nn.Linear(3, 1),  # 3 neurons to one thru second linear layer
    nn.Sigmoid(), # final activation for prediction
    )

  def forward(self, features):
    out = self.network(features)

    return out
```

```python
loss_func = nn.BCELoss()
```

```python
class myNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

Training Pipeline

```python
model = myNN(x_train.shape[1])
```

Training Pipe;ine

```python
lr = 0.1
epochs = 5

optimizer = torch.optim.SGD(model.parameters(), lr = lr)

for epoch in range(epochs):
  #forward pass
  y_pred = model(x_train.float())

  #loss calc
  loss = loss_func(y_pred, y_train.float().reshape(y_pred.shape))  #optimized loss function (no manula func for loss)
  print(loss.item())

  #backward pass
  loss.backward()

  #clear grad
  optimizer.zero_grad()

  #update para
  optimizer.step()
```

```python
# Apply a threshold to get binary predictions
y_pred_binary = (y_pred >= 0.5).float()
print("Binary Predictions:")
print(y_pred_binary)

# Calculate accuracy
accuracy = (y_pred_binary == y_test).float().mean() * 100
print(f"\nAccuracy: {accuracy.item():.2f}%")
```
