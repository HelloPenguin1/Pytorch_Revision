```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
df.drop(columns = ["id"], inplace = True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size = 0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
lencoder = LabelEncoder()
y_train = lencoder.fit_transform(y_train)
y_test = lencoder.fit_transform(y_test)
```

Numpy Array to Pytorch tensors

```python
x_train = torch.from_numpy(x_train)
x_test = torch.from_numpy(x_test)
y_train = torch.from_numpy(y_train)
y_test = torch.from_numpy(y_test)
```

```python
x_train
```

```python
x_train.shape  #30 features
```

Defining the Model

```python
class myNN():
  def __init__(self, X):
    self.weights = torch.rand(X.shape[1], 1, dtype = torch.float64, requires_grad = True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad = True) #since only one neuron in this NN

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    return torch.sigmoid(z)

  def loss_function(self, y_pred, y):
    epsilon = 1e-8
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)  #eitber 0 or 1
    return -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred)).mean()
```

```python

#imp paraT
lr = 0.1
epochs = 1
```

Training Pipeline

```python
model = myNN(x_train)
model.weights
```

```python
model.bias
```

Training Pipe;ine

```python
for epoch in range(epochs):
  #forward pass
  y_pred = model.forward(x_train)

  #loss calc
  loss = model.loss_function(y_pred, y_train)
  print(loss.item())

  #backward pass
  loss.backward()

  #update para
  with torch.no_grad():
    model.weights-= lr * model.weights.grad
    model.bias-= lr * model.bias.grad

  # zero graodents
  model.weights.grad.zero_()
  model.bias.grad.zero_()


```

```python
#evaluation

with torch.no_grad():
  y_pred = model.forward(x_test)
  encoded = torch.clamp(y_pred, min=0, max=1)
  print(encoded)
```

```python
# Apply a threshold to get binary predictions
y_pred_binary = (y_pred >= 0.5).float()
print("Binary Predictions:")
print(y_pred_binary)

# Calculate accuracy
accuracy = (y_pred_binary == y_test).float().mean() * 100
print(f"\nAccuracy: {accuracy.item():.2f}%")
```
